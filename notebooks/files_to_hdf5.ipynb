```python
import pandas as pd
from glob import glob
from joblib import Parallel, delayed
```

```python
BS_csv_path='C:/Users/Sarv/Desktop/vs_code/Git/ETL-pipeline/data/processed/balance_sheet_csv/'
PL_csv_path='C:/Users/Sarv/Desktop/vs_code/Git/ETL-pipeline/data/processed/profit_loss_csv/'
TH_csv_path='C:/Users/Sarv/Desktop/vs_code/Git/ETL-pipeline/data/processed/trade_history_csv/'

BS_csv_files=glob(BS_csv_path+'*.csv')
PL_csv_files=glob(PL_csv_path+'*.csv')
TH_csv_files=glob(TH_csv_path+'*.csv')
```

```python
def df_list_maker(file):
    df_list=[]
    try:
        df=pd.read_csv(file)
        df_list.append(df)
    except:
        pass
    return df_list

def BS_hd5_maker(df_list):
    BS_hd5=pd.HDFStore(BS_csv_path+'balance_sheet_table.h5')
    meta_df=pd.concat(df_list).reset_index(drop=True)
    BS_hd5['balance_sheet_table']=meta_df
    BS_hd5.close()
    return meta_df.shape

def PL_hd5_maker(df_list):
    PL_hd5=pd.HDFStore(PL_csv_path+'profit_loss_table.h5')
    meta_df=pd.concat(df_list).reset_index(drop=True)
    PL_hd5['profit_loss_table']=meta_df
    PL_hd5.close()
    return meta_df.shape

def TH_hd5_maker(df_list):
    TH_hd5=pd.HDFStore(TH_csv_path+'trade_history_table.h5')
    meta_df=pd.concat(df_list)
    TH_hd5['trade_history_table']=meta_df
    TH_hd5.close()
    return meta_df.shape
```

```python
# csv to h5: balance sheet
df_list=Parallel(n_jobs=-1, verbose=10000)(delayed(df_list_maker)(file) for file in BS_csv_files)
df_list=[item for sublist in df_list for item in sublist]
BS_hd5_maker(df_list)

# csv to h5: profit loss
PL_df_list=Parallel(n_jobs=-1, verbose=10000)(delayed(df_list_maker)(file) for file in PL_csv_files)
PL_df_list=[item for sublist in PL_df_list for item in sublist]
PL_hd5_maker(PL_df_list)
```

```python
#csv to h5: trade history
TH_df_list=Parallel(n_jobs=-1, verbose=500)(delayed(df_list_maker)(file) for file in TH_csv_files[:300])
TH_df_list=[item for sublist in TH_df_list for item in sublist]
TH_hd5_maker(TH_df_list)
```
